# Distribution Plots

## Correlation

```python
df.corr()
```

```python
sns.heatmap(df.corr())
```

## Scatter plot / Regression plot

```python
sns.regplot(data=df,x='tip',y='total_bill')
```

```python
sns.scatterplot()
```

## Joint Plot

A jointplot allows to study the numerical relationship between 2 numeric variables. The central chart displays the correlation. It is usually a scatterplot, a hexbin plot, a 2D histogram or a 2D density plot

```python
sns.jointplot(data=df,x=df['tip'],y=df['total_bill'],kind='hex')
```

## Pair Plot

```python
sns.pairplot(df,hue='sex')
```

```python
sns.pairplot(df,hue='size')
```

## Histogram / Distplot
