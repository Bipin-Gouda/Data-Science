```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
header={"user-agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/111.0.0.0 Safari/537.36"}
webpage=requests.get('https://www.ambitionbox.com/list-of-companies?campaign=homepage_companies_widget&page=1',headers=header).text
```

```python
soup=BeautifulSoup(webpage,'html')
```

```python
company=soup.find_all('div',class_='company-content-wrapper')     # company stores all the divs havin company content wrapper class
```

```python
len(company)    #30 companies in pg1 
```

Inspect which tag then use it

```python
name=[]
rating=[]
review=[]
companytype=[]     # prblm as 'p' tag and class_ same for companytype,HQ,Emploees,ageofcompany
HQlocn=[]
AGE=[]
no_of_employees=[]
#Tags=[[]]
#About=[[]]

for i in company:                                               #searching in those containers/'div's
    name.append(i.find('h2').text.strip())                      #find used b/c only one h2 tag in each company-content wrapper : Title
    rating.append(i.find('p',class_='rating').text.strip())
    review.append(i.find('a',class_='review-count').text.strip())
    companytype.append(i.find_all('p',class_='infoEntity')[0].text.strip())  #find_all b/c there are 4 'p' classes with class_ ='infoEntity'
    HQlocn.append(i.find_all('p',class_='infoEntity')[1].text.strip())
    AGE.append(i.find_all('p',class_='infoEntity')[2].text.strip())
    no_of_employees.append(i.find_all('p',class_='infoEntity')[3].text.strip())

d={'Name':name,'Rating':rating,'Review':review,'Companytype':companytype,'HQlocn':HQlocn,'AGE':AGE,'No_of_employees':no_of_employees}
df=pd.DataFrame(d)
```
