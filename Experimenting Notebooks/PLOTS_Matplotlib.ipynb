# PLOTS

```python
import pandas as pd
import random
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
fig,ax = plt.subplots()  # A figure containing a single axes
ax.plot([1,2,3,4],[1,4,6,3],c='red',marker="o",linestyle="dashed")
plt.xlim(1,5)
plt.ylim(1,7)
ax.set_xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Random points")
```

```python
x=[random.choice(range(1,29)) for i in range (1,50)]
y=[random.choice(range(1,41)) for i in range (1,50)]
plt.scatter(x,y,c='red')
```
